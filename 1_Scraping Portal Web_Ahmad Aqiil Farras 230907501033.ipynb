### **LIPUTAN6**

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
from datetime import datetime

headers = {"User-Agent": "Mozilla/5.0"}

# Target kategori
urls = [
    "https://www.liputan6.com/bisnis",
    "https://www.liputan6.com/tekno"
]

collected_data = []

for url in urls:
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    # Cari semua link artikel dengan pola /read/
    all_links = soup.find_all("a", href=True)
    articles = [a for a in all_links if "/read/" in a["href"]]

    print(f"Menemukan {len(articles)} artikel kandidat di {url}")

    for a in articles:
        link = a["href"]
        title = a.get_text(strip=True)

        if not title:  # skip kalau kosong
            continue

        parent = a.find_parent("article") or a.find_parent("div")

        # ringkasan
        summary_tag = parent.find("p") if parent else None
        summary = summary_tag.get_text(strip=True) if summary_tag else "Tidak ada ringkasan"

        # gambar
        img_tag = parent.find("img") if parent else None
        image_url = img_tag.get("src") if img_tag else "Tidak ada gambar"

        # waktu (kadang hanya ada di halaman detail, jadi opsional)
        time_tag = parent.find("time") if parent else None
        publication_time = time_tag.get_text(strip=True) if time_tag else "Tidak ada waktu"

        collected_data.append({
            "title": title,
            "link": link,
            "summary": summary,
            "image_url": image_url,
            "publication_time": publication_time,
            "category": "Bisnis" if "bisnis" in url else "Tekno",
            "url_source": url,
            "scraped_at": datetime.now().isoformat()
        })

# EXPORT
if collected_data:
    file_name_base = "list_post_liputan6_bisnis_tekno"
    df = pd.DataFrame(collected_data)

    with open(f"{file_name_base}.json", "w", encoding="utf-8") as f:
        json.dump({"list_post": collected_data}, f, ensure_ascii=False, indent=4)

    df.to_csv(f"{file_name_base}.csv", index=False, encoding="utf-8")
    df.to_excel(f"{file_name_base}.xlsx", index=False)

    print(f" Berhasil scraping {len(collected_data)} artikel dari Liputan6 Bisnis & Tekno")
else:
    print(" Tidak ada artikel ditemukan.")
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json
from datetime import datetime

headers = {"User-Agent": "Mozilla/5.0"}

# Target kategori
urls = [
    "https://www.antaranews.com/ekonomi",
    "https://www.antaranews.com/tekno"
]

collected_data = []

for url in urls:
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    # Ambil semua link <a> dengan pola /berita/
    all_links = soup.find_all("a", href=True)
    articles = [a for a in all_links if "/berita/" in a["href"]]

    print(f"Menemukan {len(articles)} artikel kandidat di {url}")

    for a in articles:
        link = a["href"]
        title = a.get_text(strip=True)

        if not title:  # skip kalau kosong
            continue

        parent = a.find_parent("article") or a.find_parent("div")

        # ringkasan
        summary_tag = parent.find("p") if parent else None
        summary = summary_tag.get_text(strip=True) if summary_tag else "Tidak ada ringkasan"

        # gambar
        img_tag = parent.find("img") if parent else None
        image_url = img_tag.get("src") if img_tag else "Tidak ada gambar"

        # waktu
        time_tag = parent.find("time") if parent else None
        publication_time = time_tag.get_text(strip=True) if time_tag else "Tidak ada waktu"

        collected_data.append({
            "title": title,
            "link": link,
            "summary": summary,
            "image_url": image_url,
            "publication_time": publication_time,
            "category": "Ekonomi" if "ekonomi" in url else "Tekno",
            "url_source": url,
            "scraped_at": datetime.now().isoformat()
        })

# EXPORT
if collected_data:
    file_name_base = "list_post_antara_fallback"
    df = pd.DataFrame(collected_data)

    with open(f"{file_name_base}.json", "w", encoding="utf-8") as f:
        json.dump({"list_post": collected_data}, f, ensure_ascii=False, indent=4)

    df.to_csv(f"{file_name_base}.csv", index=False, encoding="utf-8")
    df.to_excel(f"{file_name_base}.xlsx", index=False)

    print(f"✅ Berhasil scraping {len(collected_data)} artikel dari Antara News (fallback)")
else:
    print("⚠️ Tidak ada artikel ditemukan.")
```
